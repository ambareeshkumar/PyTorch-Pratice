#### Saving & Loading Models Recommended Way

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class SampleNeuralNet(nn.Module):

    def __init__(self, n_features):
        super(SampleNeuralNet, self).__init__()
        self.fc1 = nn.Linear(n_features, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, 1)
    
    def forward(self, x):
        self.fc1_out = F.relu(self.fc1(x))
        self.fc2_out = F.relu(self.fc2(self.fc1_out))
        self.fc3_out = self.fc3(self.fc2_out)
        return self.fc3_out

model = SampleNeuralNet(10)
```

```python
# Saving the model
File = 'Model/model_new_save.pth'

torch.save(model.state_dict(), File)
```

```python
# loading the saved model
load_model = SampleNeuralNet(10)
load_model.load_state_dict(torch.load(File))

load_model.eval()

for param in load_model.parameters():
    print(param)
```
